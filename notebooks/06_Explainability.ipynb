## Spectral Band Importance

The multispectral model benefits from Sentinel-2 spectral bands.

Important bands:

### B8 (Near Infrared)
Useful for vegetation monitoring.

### B5-B7 (Red Edge)
Capture plant health and chlorophyll information.

### B11-B12 (SWIR)
Useful for moisture and soil analysis.

These bands help distinguish agricultural classes that appear visually similar in RGB imagery.
