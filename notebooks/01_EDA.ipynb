```python
rgb_train = "../data/EuroSAT/train"

print(os.listdir(rgb_train))
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import os

class_name = os.listdir(rgb_train)[0]

img_path = os.path.join(
    rgb_train,
    class_name,
    os.listdir(os.path.join(rgb_train,class_name))[0]
)

print(img_path)

img = Image.open(img_path)

print("Size:", img.size)

plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
import os

ms_train = "../data/EuroSATallBands/train"

print(os.listdir(ms_train))
```

```python
class_name = os.listdir(ms_train)[0]

sample_tif = os.path.join(
    ms_train,
    class_name,
    os.listdir(os.path.join(ms_train, class_name))[0]
)

print(sample_tif)
```

```python
import rasterio

with rasterio.open(sample_tif) as src:
    img = src.read()

print(img.shape)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(3,5, figsize=(14,8))

for i in range(13):
    ax = axes.flatten()[i]
    ax.imshow(img[i], cmap="gray")
    ax.set_title(f"Band {i+1}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```
